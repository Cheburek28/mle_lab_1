```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../data/data.csv')
data.info()
```

```python
data['diagnosis'].replace({'M':1, 'B':0}, inplace = True)
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.drop(['id','Unnamed: 32'], axis = 1)
data
```

```python
y = data["diagnosis"]
X = data.drop("diagnosis", axis=1 )

y.shape, X.shape
```

```python
# Посмотрим на корреляцию, между признаками и диагнозами
import seaborn as sns
import matplotlib.pyplot as plt

corr = data.corr()

sns.clustermap(corr, annot = True, figsize = (18, 15))
plt.show()
```

```python
# Откинем всех у кого корреляция меньше 70%
CORR_THRESHOLD = 0.7


corr_filter = np.abs(corr["diagnosis"]) > CORR_THRESHOLD

feautures_names = corr.columns[corr_filter].to_list()
feautures_names
```

```python
y = data["diagnosis"]
X = data[feautures_names].drop(["diagnosis"], axis = 1)

y.shape, X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 42)
```

```python
from sklearn.svm import SVC

clf_svm= SVC() # clasifying model
clf_svm.fit(X_train,y_train) # fit this model on training dataset

acc_svm = clf_svm.score(X_test, y_test) # rounding upto 2 decimal places by multiplying it by 100
print(acc_svm)
```
